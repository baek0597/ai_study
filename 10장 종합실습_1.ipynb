```python
# train 하위 폴더별 이미지 갯수 파악
for lb in labels:
    print(lb, len(os.listdir(train_dir+lb)))
```

```python
# glob로 이미지데이터 경로 저장하기
paths = glob(train_dir+"*/*")
len(paths)
```

```python
# 이미지 10개 출력
paths[0:5]
```

```python
# 첫번째 이미지 불러와서 출력하기
image_pil = Image.open(paths[0])
image_pil
```

```python
# array로 변환
image = np.array(image_pil)
print(image.shape)
image
```

```python
# array 를 이미지로 출력
plt.imshow(image,"gray")
plt.show()
```

```python
# 전체 이미지 사이즈 저장
heights = []
widths = []

for path in paths:
    image_pil = Image.open(path)
    image = np.array(image_pil)
    h, w = image.shape
    
    heights.append(h)
    widths.append(w)
```

```python
# 전체 이미지 사이즈 그려보기
plt.figure(figsize=(20, 10))

plt.subplot(121)
plt.hist(heights)
plt.title('Heights')


plt.subplot(122)
plt.hist(widths)
plt.title('Widths')


plt.show()
```

```python
# 텐서플로우로 이미지 불러오기
gfile = tf.io.read_file(paths[0])
image = tf.io.decode_image(gfile, dtype=tf.float32)
```

```python
image.shape # 이미지 사이즈
```

```python
# 이미지 출력
plt.imshow(image,"gray")
plt.show()
```

```python
# 텐서플로우로 데이터 제너레이션 
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
image.shape
```

```python
inputs = image[tf.newaxis, ...]  #배열 추가하기
inputs.shape
```

```python
# 데이터 제너레이터 생성
datagen = ImageDataGenerator(
    rescale = 1. /255.,
    zoom_range=0.2,
    rotation_range=1,
    shear_range=0.5,
    brightness_range= (0,2)
    
)

result = next(iter(datagen.flow(inputs)))  #실행


plt.subplot(121)
plt.title("Original Image")
plt.imshow(inputs[0, :, :, 0], 'gray')
plt.subplot(122)
plt.title("Transformed Image")
plt.imshow(result[0, :, :, 0], 'gray')
plt.show()
```
